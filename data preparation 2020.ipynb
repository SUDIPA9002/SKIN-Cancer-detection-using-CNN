```python
import pandas as pd
```

```python
df = pd.read_csv('ISIC_2020_Training_GroundTruth.csv')
```

```python
df.tail()
```

```python
df.benign_malignant.replace(('benign', 'malignant'), (1, 0), inplace=True)
```

```python
df.head()
```

```python
benign_list = df[df['benign_malignant']==1]['image_name'].tolist()
len(benign_list)
```

```python
import os
import shutil

source = 'D:/ISIC_2020_Training_Input/ISIC_2020_Training_Input/'
destination = 'D:/ISIC_2020_Training_Input/Benign/' 
allfiles = os.listdir(source)
  
for f in allfiles:
    m=f.split('.')[0]
    if m in benign_list:
        file="benign_" + str(f)
        print(file)
        shutil.copy(source + f, destination + file)
```

```python
malignant_list = df[df['benign_malignant']==0]['image_name'].tolist()
len(malignant_list)
```

```python
source = 'D:/ISIC_2020_Training_Input/ISIC_2020_Training_Input/'
destination = 'D:/ISIC_2020_Training_Input/Malignant/'
  
allfiles = os.listdir(source)
  
for f in allfiles:
    m=f.split('.')[0]
    if m in malignant_list:
        file="malignant_" + str(f)
        
        shutil.copy(source + f, destination + file)
```
